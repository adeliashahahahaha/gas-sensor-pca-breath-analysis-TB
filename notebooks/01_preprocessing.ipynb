```python
# preprocessing data
import pandas as pd
import os

os.makedirs('data', exist_ok=True)

for i in range(1, 11):

    input_file = f'raw/batch{i}.dat'
    output_file = f'data/batch{i}_clean.csv'

    if not os.path.exists(input_file):
        print(f'❌ {input_file} tidak ditemukan')
        continue

    data = []

    with open(input_file, 'r') as f:
        lines = f.readlines()

    for line in lines:
        parts = line.strip().split()

        # ambil label dan concentration
        first_part = parts[0]

        if ';' in first_part:
            label, concentration = first_part.split(';')
            label = int(label)
            concentration = float(concentration)
            feature_items = parts[1:]
        else:
            label = int(first_part)
            concentration = None
            feature_items = parts[1:]

        features = []

        for item in feature_items:
            if ':' in item:
                value = float(item.split(':')[1])
                features.append(value)

        if len(features) == 128:
            row = [label, concentration] + features
            data.append(row)

    columns = ['label', 'concentration_ppmv'] + [
        f'feature_{j}' for j in range(1, 129)
    ]

    df = pd.DataFrame(data, columns=columns)

    df.to_csv(output_file, index=False)

    print(f'✅ batch{i} berhasil dikonversi')
    print(df.head())
    print(df.shape)
    print('-' * 50)
```
